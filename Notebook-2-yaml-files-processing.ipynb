```python
dictmatchnum=dict()
dictmatchnum[2008]=0
dictmatchnum[2009]=0
dictmatchnum[2010]=0
dictmatchnum[2011]=0
dictmatchnum[2012]=0
dictmatchnum[2013]=0
dictmatchnum[2014]=0
dictmatchnum[2015]=0
dictmatchnum[2016]=0
dictmatchnum[2017]=0
dictmatchnum[2018]=0
dictmatchnum[2019]=0
```

```python
import glob
#for file in glob.glob("D:\\Learning\\Matches_IPL\\ipl_files\\*.yaml"):
#   print(file)
```

```python
count=0
file_num=0
for file in glob.glob("D:\\Learning\\Matches_IPL\\ipl_files\\*.yaml"):
    count=count+1
    #f=open('335982.yaml','r')
    file_num=file_num+1
    print(str(file_num), ' - ',file)
    f=open(file,'r')
    doc = yaml.load(f)

    dictrow=dict()
    try:
        match_year=doc['info']['dates'][0].year
        match_number=dictmatchnum[doc['info']['dates'][0].year]+1
        dictmatchnum[match_year]=match_number
    except:
        match_year=int(doc['info']['dates'][0].split('-')[0])
        match_number=dictmatchnum[match_year]+1
        dictmatchnum[match_year]=match_number
    total_5=0
    total=0
    wickets_1=0
    for i in range(len(doc['innings'][0]['1st innings']['deliveries'])):
        for j in doc['innings'][0]['1st innings']['deliveries'][i].keys():
            if(j<5.1):
                total_5=total_5+doc['innings'][0]['1st innings']['deliveries'][i][j]['runs']['total']
                total=total+doc['innings'][0]['1st innings']['deliveries'][i][j]['runs']['total']
                if ('wicket' in doc['innings'][0]['1st innings']['deliveries'][i][j].keys()):
                    wickets_1=wickets_1+1
            else:
                total=total+doc['innings'][0]['1st innings']['deliveries'][i][j]['runs']['total']
    #len(doc['innings'][0]['1st innings']['deliveries'])
    #doc['innings'][0]['1st innings']['deliveries']
    if(doc['info']['toss']['decision']!='field'):
        dictrow['Toss_Won']=1
    else:
        dictrow['Toss_Won']=0
    dictrow['5_Over_score']=total_5
    dictrow['5_Over_Wicket']=wickets_1
    dictrow['Total_Score']=total
    dictrow['Match_Year']=match_year
    dictrow['Match_Number'] = match_number
    try:
        dictrow['Match_City'] = doc['info']['city']
    except:
        dictrow['Match_City'] ='Unknown'
    dictrow['Match_Stadium'] = doc['info']['venue']
    dictrow['Match_Teams'] = doc['info']['teams'][0] + ' - ' + doc['info']['teams'][1]
    dictrow['First_Innings']=1
    dictrow['Team']= doc['innings'][0]['1st innings']['team']
    #print(count)
    dfrow=pd.DataFrame(dictrow,index=[count])
    df=df.append(dfrow)

    total_5_2=0
    total_2=0
    wickets_2=0
    dictrow=dict()
    count=count+1
    for i in range(len(doc['innings'][1]['2nd innings']['deliveries'])):
        for j in doc['innings'][1]['2nd innings']['deliveries'][i].keys():
            if(j<5.1):
                total_5_2=total_5_2+doc['innings'][1]['2nd innings']['deliveries'][i][j]['runs']['total']
                total_2=total_2+doc['innings'][1]['2nd innings']['deliveries'][i][j]['runs']['total']
                if ('wicket' in doc['innings'][1]['2nd innings']['deliveries'][i][j].keys()):
                    wickets_2=wickets_2+1
            else:
                total_2=total_2+doc['innings'][1]['2nd innings']['deliveries'][i][j]['runs']['total']
    if(doc['info']['toss']['decision']=='field'):
        dictrow['Toss_Won']=1
    else:
        dictrow['Toss_Won']=0
    dictrow['5_Over_score']=total_5_2
    dictrow['5_Over_Wicket']=wickets_2
    dictrow['Total_Score']=total_2
    #dictrow['Match_Year']= doc['info']['dates'][0].year
    dictrow['Match_Year']=match_year
    dictrow['Match_Number'] = match_number
    try:
        dictrow['Match_City'] = doc['info']['city']
    except:
        dictrow['Match_City'] ='Unknown'
    dictrow['Match_Stadium'] = doc['info']['venue']
    dictrow['Match_Teams'] = doc['info']['teams'][0] + ' - ' + doc['info']['teams'][1]
    dictrow['First_Innings']=0
    dictrow['Team']= doc['innings'][1]['2nd innings']['team']
    #print(count)
    dfrow=pd.DataFrame(dictrow,index=[count])
    df=df.append(dfrow)
    #len(doc['innings'][0]['1st innings']['deliveries'])
    #doc['innings'][0]['1st innings']['deliveries']
```
