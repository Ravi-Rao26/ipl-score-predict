```python
###  FUntion to define dictionary
def Year_Dict_Teams(datadf,year, Inns):
    dictteam={}
    for i in range(len(datadf)):
        myear=datadf.loc[i:i, 'Match_Year'].values[0]
        mnum=datadf.loc[i:i, 'Match_Number'].values[0]
        minns= datadf.loc[i:i, 'First_Innings'].values[0]
        tc=datadf.loc[i:i, 'Team_Codes'].values[0]
        #print(myear)
        if(myear==year and minns==Inns ):
            if(tc in dictteam):
                dictteam[tc].append(mnum)
            else:
                dictteam[tc]=[mnum]
    return(dictteam)
```

```python
import numpy as np
def get_Last_match_num(dict1,matchnum,tm):
    matchlist=dict1[tm]
    previous_matches=[]
    for m in matchlist:
        if(matchnum > m):
            previous_matches.append(m)
    if(len(previous_matches)!=0):
        return(np.max(previous_matches))
    else:
        return('Not Available')
    
    
get_Last_match_num(dict_2008_First_Inn,1,'KKR')
```

```python
###  FUntion to define dictionary
def Year_Dict(datadf,year):
    dictteam={}
    for i in range(len(datadf)):
        myear=datadf.loc[i:i, 'Match_Year'].values[0]
        mnum=datadf.loc[i:i, 'Match_Number'].values[0]
        #minns= datadf.loc[i:i, 'First_Innings'].values[0]
        tc=datadf.loc[i:i, 'Team_Codes'].values[0]
        #print(myear)
        if(myear==year):
            if(tc in dictteam):
                dictteam[tc].append(mnum)
            else:
                dictteam[tc]=[mnum]
    return(dictteam)
```
